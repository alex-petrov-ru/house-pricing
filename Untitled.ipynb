# Подготовка данных: заполнение пропусков


Есть несколько классических подходов - дропнуть строки с такими данными, заполнить средним, заполнить чем-то логичным (в зависимости от специфики данных), построить, например,RF и заполнять пропуски итеративно. Для начала исследуем пропущенные значения

```python
total = df_train.isnull().sum().sort_values(ascending=False)
percent = (df_train.isnull().sum()/df_train.isnull().count()
           ).sort_values(ascending=False)
missing_data = pd.concat([total, percent], axis=1, keys=['Total', 'Pecent'])
missing_data.head(20)
```

Давайте проанализируем: первые 6 кандидатов имеют большой процент пропущенных значений (больше 17) - так как эти признаки не имели сильной корреляции по предыдущему анализу - заменим на наиболее частое значение По остальным удалим пропущеннные значения

```python
x_train = x_train.drop(missing_data[missing_data['Total'] > 81].index, 1)
```

```python
x_train = x_train.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df_test = df_test.drop(missing_data[missing_data['Total'] > 81].index, 1)
df_test = df_test.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
from sklearn.preprocessing import LabelEncoder

cols = x_train.select_dtypes(include='object').columns

for col in cols:
    lbl = LabelEncoder()
    lbl.fit(list(x_train[col].values))
    x_train[col] = lbl.transform(list(x_train[col].values))
    df_test[col] = lbl.transform(list(df_test[col].values))

print(f'Shape all_data: {x_train.shape}')
```

# GridSearch

Запустим GridSearch чтобы подобрать параметры получше.

Что будем настраивать: параметры max_depth, min_child_weight и gamma непосредственно ограничивают сложность модели, subsample и colsample_bytree делают её более устойчивой к шуму за счет добавления случайного выбора наблюдений и предикторов. reg_lambda и reg_alpha - параметры регуляризации: увеличивая можно сделать модель более устойчивой.

```python
from sklearn.model_selection import GridSearchCV

params = {'min_child_weight': [4, 5],
          'gamma': [i/10 for i in range(3, 6)],
          'subsample': [i/10 for i in range(6, 11)],
          'colsample_bytree': [i/10 for i in range(4, 11)],
          'max_depth': [2, 3, 4],
          'reg_lambda': [i/10 for i in range(7, 9)],
          'reg_alpha': [i/10 for i in range(4,7)]}

model = xgb.XGBRegressor(nthread=-1)

grid = GridSearchCV(model, params)
grid.fit(x_train, y_train)
```
